```python
%%time
#infiles='/home/agopal/storage/ihesp/HRMIP/*.nc'
infiles='/ihesp/archive/HRMIP/B.E.13.B1950C5.ne120_t12.cesm-ihesp-1950cntl.013/ocn/hist/daily_merged/B.E.13.B1950C5.ne120_t12.cesm-ihesp-1950cntl.013.pop.h.0070-*.nc'
#ds = xr.open_mfdataset(infiles,combine='nested',data_vars=['SST'],concat_dim='time',chunks={'nlat':100,'nlon':100,'z_t':1})
#ds = xr.open_mfdataset(infiles,combine='by_coords',data_vars=['SST'],concat_dim='time',parallel='True',chunks={'nlat':-1,'nlon':-1,'z_t':1})
#ds = xr.open_mfdataset(infiles,combine='by_coords',concat_dim='time',parallel='True',chunks={'nlat':-1,'nlon':-1,'z_t':1,''})
with performance_report(filename="dask-report-open.html"):
    ds = xr.open_mfdataset(infiles,combine='by_coords',concat_dim='time',parallel='True')


ds

# size of data in MB
#ds.nbytes/1e6
```

```python
#ds = ds.sel(nlat=slice(0, 600),nlon=slice(0, 600))
#ds = ds.sel(z_t=slice(0,0))
#ds
rechunked = ds.chunk({'nlat':-1,'nlon':-1,'z_t':1,'time':3})
rechunked
```

```python
#temp_clim = ds.TEMP.groupby('time.month').mean(dim='time')
sst_clim = rechunked.SST.groupby('time.month').mean(dim='time')
```
