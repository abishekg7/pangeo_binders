```python
infile='/scratch/training/CESM_post/datasets/hires/B.E.13.B1950C5.ne120_t12.cesm-ihesp-1950cntl.013.pop.h.0060-01.nc'
ds = xr.open_dataset(infile)    
#ds = xr.open_dataset(infile,chunks={'time':1,'nlat':400,'nlon':600})    
ds
```

```python
fig = plt.figure(figsize=(14,6))
ax = plt.axes(projection=cartopy.crs.PlateCarree());

ax.pcolormesh(ds.TLONG,ds.TLAT,sst_0,transform=cartopy.crs.PlateCarree(),cmap='jet');
```

```python
POPgridsFile_tx01v2 = 'POPgrids_tx0.1v2_2020Jun29.nc'
ds_grid_tx01v2 = xr.open_dataset(POPgridsFile_tx01v2)
ds_grid_tx01v2.
```

```python
#Fix by Dapeng Li
sst_HR_raw = ds.SST.isel(time=0)
sst_HR_raw2 = sst_HR_raw.where(ds.KMT>0,np.nan) # use ds_HR.KMT to mask land with NaN


fig = plt.figure(figsize=(14,6))
ax = plt.axes(projection=cartopy.crs.PlateCarree());
ax.pcolormesh(ds_grid_tx01v2.TLONG, ds_grid_tx01v2.TLAT, sst_HR_raw2,transform=cartopy.crs.PlateCarree(),cmap='jet')
```

```python
fig = plt.figure(figsize=(14,6))
ax = plt.axes(projection=cartopy.crs.PlateCarree());
range_lat=slice(200,1200)
range_lon=slice(500,2000)
lon = ds_grid_tx01v2.TLONG.isel(nlat=range_lat,nlon=range_lon)

lat = ds_grid_tx01v2.TLAT.isel(nlat=range_lat,nlon=range_lon)
var = sst_HR_raw2.isel(nlat=range_lat,nlon=range_lon)
ax.pcolormesh(lon,lat,var,transform=cartopy.crs.PlateCarree(),cmap='jet');
```
