```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import intake


xr.set_options(display_style='html')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
cat_url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"
col = intake.open_esm_datastore(cat_url)
col
```

```python
cat = col.search(experiment_id=['historical'], table_id='Oyr', variable_id='o2',
                 grid_label='gn')

cat.df
```

```python
cat = col.search(activity_id='CMIP', table_id='Amon', variable_id='tas', experiment_id='historical',
                  institution_id="NCAR", source_id="CESM2", member_id="r10i1p1f1")
cat.df
```

```python
dset_dict = cat.to_dataset_dict(zarr_kwargs={'consolidated': True})
list(dset_dict.keys())
```

```python
ds = dset_dict['CMIP.NCAR.CESM2.historical.Amon.gn']
ds
```

```python
import cartopy
#ds.tas.sel(time='1950-01').squeeze().plot()
fig = plt.figure(figsize=(14,6))
ax = plt.axes(projection=cartopy.crs.PlateCarree());

ax.pcolormesh(ds['lon'], ds['lat'], ds.tas.sel(time='1950-01'),
                   cmap='jet')
#plt.show()
```

```python
files_area = col.search(variable_id='areacella', activity_id='CMIP',
                         experiment_id='historical', institution_id="NCAR", source_id="CESM2")

dset_dict = files_area.to_dataset_dict()
list(dset_dict.keys())
#ds_area = xr.open_dataset(files_area[0])
#ds_area
```

```python
ds_area = dset_dict['CMIP.NCAR.CESM2.historical.fx.gn']
ds_area
```

```python
total_area = ds_area.areacella.sum(dim=['lon', 'lat'])
ta_timeseries = (ds.tas * ds_area.areacella).sum(dim=['lon', 'lat']) / total_area
ta_timeseries
```

```python
ta_timeseries.plot(label='monthly')
ta_timeseries.rolling(time=12).mean().plot(label='12 month rolling mean')
plt.legend()
plt.title('Global Mean Surface Air Temperature')
```
